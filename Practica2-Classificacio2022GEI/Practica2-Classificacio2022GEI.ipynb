```python
x_t, x_v, y_t, y_v = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
##########LOGISTIC REGRESSION###########
```

```python
from sklearn.linear_model import LogisticRegression

LogisticRegression = LogisticRegression(C=0.5)
LogisticRegression.fit(x_t, y_t)
preds = LogisticRegression.predict(x_v)

#calcular la f1 score
score = f1_score(y_v, preds)
print("f1_score = " + str(score))

#confusion matrix
cm = metrics.confusion_matrix(y_v, preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp = disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
##################   SVM   ##################
```

```python
from sklearn.svm import SVC
C = 9.0
gamma = 0.2
degree = 8.0
kernel_options = ['linear', 'poly', 'sigmoid', 'rbf']
for ker in kernel_options:
    if ker == 'poly':
        svclassifier = svm.SVC(kernel=ker, degree=degree, gamma='auto', C=C)
    elif ker == 'linear':        
        svclassifier = svm.LinearSVC(C=C, max_iter=1000000)
    else:
        svclassifier = svm.SVC(kernel='rbf', gamma=gamma, C=C)
        
    svclassifier.fit(x_t, y_t)
    svcPreds = svclassifier.predict(x_v)
    print('SVM amb kernel ' + ker)
    #mirar la f1 score
    score = f1_score(y_v, svcPreds)
    print(score)
    #confusion matrix
    cm = metrics.confusion_matrix(y_v, svcPreds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp = disp.plot(cmap=plt.cm.Blues)
    plt.show()            
```

```python
##############  KNN  ###################
    # TOTES LES k TORNEN LA MATEIXA ACCURACY
```

```python
from sklearn.neighbors import KNeighborsClassifier

for i in range(5)[1:]:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_t, y_t)
    preds = svclassifier.predict(x_v)
    print('KNN amb k ' + str(i))
    #mirar la f1 score
    score = f1_score(y_v, preds)
    print(score)
    #confusion matrix
    cm = metrics.confusion_matrix(y_v, preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp = disp.plot(cmap=plt.cm.Blues)
    plt.show()
```

```python
##############  PERCEPTRÓ  ###################
```

```python
from sklearn.linear_model import Perceptron
clf = Perceptron(max_iter=1000, eta0=0.1, tol=1e-3, random_state=26)
clf.fit(x_t, y_t)
preds = clf.predict(x_v)
#mirar la f1 score
score = f1_score(y_v, preds)
print(score)
#confusion matrix
cm = metrics.confusion_matrix(y_v, preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp = disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
##############  ENSEMBLES  ###################
```

```python
##############  RANDOM FOREST  ###################
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, criterion='log_loss')
rf.fit(x_t, y_t)
preds = rf.predict(x_v)
#mirar la f1 score
score = f1_score(y_v, preds)
print(score)
#confusion matrix
cm = metrics.confusion_matrix(y_v, preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp = disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
##############  BAGGING  ###################
from sklearn.ensemble import BaggingClassifier

##BAGGING DE DECISION TREES:
clf = BaggingClassifier(n_estimators=43, random_state=28)
clf.fit(x_t, y_t)
preds = clf.predict(x_v)
#mirar la f1 score
score = f1_score(y_v, preds)
print('Bagging de DecisionTrees: '+ str(score))
#confusion matrix
cm = metrics.confusion_matrix(y_v, preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp = disp.plot(cmap=plt.cm.Blues)
plt.show()

##BAGGING DE SVC:
clf = BaggingClassifier(n_estimators=43, random_state=28, base_estimator=SVC())
clf.fit(x_t, y_t)
preds = clf.predict(x_v)
#mirar la f1 score
score = f1_score(y_v, preds)
print('Bagging de SVC: '+ str(score))
#confusion matrix
cm = metrics.confusion_matrix(y_v, preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp = disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
##############  BOOSTING  ###################
from sklearn.ensemble import AdaBoostClassifier
clf = AdaBoostClassifier(n_estimators=100)
clf.fit(x_t, y_t)
preds = clf.predict(x_v)
#mirar la f1 score
score = f1_score(y_v, preds)
print('AdaBoost: '+ str(score))
#confusion matrix
cm = metrics.confusion_matrix(y_v, preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp = disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
# Take all features except target on X and target which is potability on Y.
X = df.values[:, :9]
y = df.values[:,9]

from sklearn.preprocessing import label_binarize
# Use label_binarize to be multi-label like settings
Y = label_binarize(y, classes=[0, 1])
n_classes = 2

fig, sub = plt.subplots(1, 2, figsize=(16,6))
sub[0].scatter(X[:,0], y, c=y, cmap=plt.cm.coolwarm, edgecolors='k')
sub[1].scatter(X[:,1], y, c=y, cmap=plt.cm.coolwarm, edgecolors='k')
sub[0].set_xlabel('ph')
sub[0].set_ylabel('Potability')
sub[1].set_xlabel('Hardness')
sub[1].set_ylabel('Potability')

fig, sub = plt.subplots(1, 2, figsize=(16,6))
sub[0].scatter(X[:,2], y, c=y, cmap=plt.cm.coolwarm, edgecolors='k')
sub[1].scatter(X[:,3], y, c=y, cmap=plt.cm.coolwarm, edgecolors='k')
sub[0].set_xlabel('Solids')
sub[0].set_ylabel('Potability')
sub[1].set_xlabel('Chloramines')
sub[1].set_ylabel('Potability')

fig, sub = plt.subplots(1, 2, figsize=(16,6))
sub[0].scatter(X[:,4], y, c=y, cmap=plt.cm.coolwarm, edgecolors='k')
sub[1].scatter(X[:,5], y, c=y, cmap=plt.cm.coolwarm, edgecolors='k')
sub[0].set_xlabel('Sulfates')
sub[0].set_ylabel('Potability')
sub[1].set_xlabel('Conductivity')
sub[1].set_ylabel('Potability')

fig, sub = plt.subplots(1, 2, figsize=(16,6))
sub[0].scatter(X[:,6], y, c=y, cmap=plt.cm.coolwarm, edgecolors='k')
sub[1].scatter(X[:,7], y, c=y, cmap=plt.cm.coolwarm, edgecolors='k')
sub[0].set_xlabel('Organic_carbon')
sub[0].set_ylabel('Potability')
sub[1].set_xlabel('Trihalomethanes')
sub[1].set_ylabel('Potability')

fig, sub = plt.subplots(figsize=(16,6))
sub.scatter(X[:,8], y, c=y, cmap=plt.cm.coolwarm, edgecolors='k')
sub.set_xlabel('Turbidity')
sub.set_ylabel('Potability')


particions = [0.5, 0.7, 0.8]

for part in particions:
    x_t, x_v, y_t, y_v = train_test_split(X, Y, train_size=part)
    
    y_t = y_t.ravel()
    #Creem el regresor logístic
    logireg = LogisticRegression(C=2.0, fit_intercept=True, penalty='l2', tol=0.001)

    # l'entrenem
    logireg.fit(x_t, y_t)

    print ("Correct classification Logistic ", part, "% of the data: ", logireg.score(x_v, y_v))
    
    #Creem el regresor logístic
    svc = svm.SVC(C=10.0, kernel='rbf', gamma=0.9, probability=True)

    # l'entrenem 
    svc.fit(x_t, y_t)
    probs = svc.predict_proba(x_v)
    print ("Correct classification SVM      ", part, "% of the data: ", svc.score(x_v, y_v))
print(np.shape(probs))
```

Tal com podeu llegir a [l'API de sklearn](http://scikit-learn.org/stable/auto_examples/model_selection/plot_precision_recall.html), en comptes de fer una corba per cada classe, podem considerar totes les classes en conjunt en una sola corba (1 si hem predit la classe correcta, 0 si no). Això es coneix com a `micro-averaging`. 

Així, veureu que la funció `f1_score` utilitza el paràmetre `macro` per calcular la precision-recall-f1 per clase, i després fer la mitja pr a totes les classes; i `micro` per utilitzar totes les prediccions (i errors de FN, FP) per a calcular una única precision-recall-f1 per a totes les classes juntes.

Si voleu calcular la corba Precision-Recall quan utilitzeu el K-fold, cal calcular les corbes per a cada fold i després [fer la mitja de tots els folds per obtenir la corba final](https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc_crossval.html#sphx-glr-auto-examples-model-selection-plot-roc-crossval-py). En el cas del LOOCV no té sentit fer la mitja la corba PR perquè hauriem de fer servir totes les mostres com a $y$.
